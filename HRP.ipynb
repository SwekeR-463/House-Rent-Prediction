```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\RYZEN-5\OneDrive\Desktop\Machine Learning\House Rent Prediction\House_Rent_Dataset.csv')
```

```python
X = df[['Size']]
y = df['Rent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
linr = LinearRegression()
```

```python
linr.fit(X_train, y_train)
```

```python
predictions = linr.predict(X_test)
```

```python
plt.scatter(df["Size"], df["Rent"])
plt.xlabel("Size")
plt.ylabel("Rent")
```

```python
plt.scatter(X_train, y_train, alpha=0.7, label='Training data', color='blue')
plt.scatter(X_test, y_test, alpha=0.7, label='Testing Data', color='black')
plt.xlabel('Size')
plt.ylabel('Rent')
plt.legend()
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, predictions, color='blue', linewidth=2)
plt.xlabel('Size')
plt.ylabel('Rent')
plt.title('Linear Regression for House Rent Prediction')
plt.show()
```

```python
result_df = pd.DataFrame({'Size': X_test['Size'].values, 'Actual Rent': y_test.values, 'Predicted Rent': predictions})
print(result_df)
```
